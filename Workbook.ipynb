```python
#### Importing package dependencies
import os, datetime, IPython.display, time
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from utils import WindowGenerator, lstm

# mpl.rcParams['figure.figsize'] = (8, 6)
# mpl.rcParams['axes.grid'] = False
```

```python
df = pd.read_csv('hourly.csv',index_col=0)
```

```python
### Split testing and training data

column_indices = {name: i for i, name in enumerate(df.columns)}

n = len(df)
train_df = df[(df['year']==2004) | (df['year']==2005)]
val_df = df[df['year']==2006]
test_df = df[df['year']==2007]
```

```python
input_width = 168
label_width = 72
data_window = WindowGenerator(input_width=input_width,
                               label_width=label_width,
                               train_df = train_df[['Number of Patients']], val_df=val_df[['Number of Patients']], 
                               test_df=test_df[['Number of Patients']],label_columns=['Number of Patients'])
```

```python
model_3day = lstm(input_width = input_width,label_width = label_width,input_features = data_window.example[0].shape[2])
```

```python
start = time.time()
history = model_3day.fit(data_window.train,data_window.val)
end = time.time()
IPython.display.clear_output()
print('Training time:', end-start, 'seconds')
```

```python
### Plot the training curve

plt.plot(history.history['mean_absolute_error'])
plt.plot(history.history['val_mean_absolute_error'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
model_3day.load_weights('checkpoints/best_model_numfeatures='+str(data_window.example[0].shape[2])+' output_width=' +str(label_width) +'.ckpt'
)
```

```python
performance = {}
performance['LSTM 3day'] = model_3day.evaluate(data_window.test)
```

```python
start = time.time()
y_LSTM_3day = model_3day.predict(data_window.test)
end = time.time()
print('Prediction time:', end - start, 'seconds')
np.savetxt('forecasts/y_LSTM_3day.txt', y_LSTM_3day)
```
